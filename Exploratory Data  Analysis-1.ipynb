# Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.
Ans:The wine quality dataset is a well-known dataset that contains information about the physicochemical properties of different types of wine. It has 12 features, including 11 input variables and 1 output variable (quality).

The following are the key features of the wine quality dataset and their importance in predicting the quality of wine:

1-Fixed acidity: It refers to the amount of fixed acids in wine, which are usually tart and contribute to its taste. Fixed acidity plays a crucial role in determining the wine's pH, which in turn affects its taste, color, and stability.

2-Volatile acidity: It is the measure of the amount of volatile acids in wine, which are responsible for giving it a vinegar-like taste. High levels of volatile acidity can negatively impact wine quality.

3-Citric acid: It is an organic acid that adds a citrusy flavor to wine. Citric acid can help balance out other acids in wine and give it a refreshing taste.

4-Residual sugar: It refers to the amount of sugar remaining in wine after fermentation. It can influence the sweetness and body of wine and affect its overall taste.

5-Chlorides: It is a measure of the amount of salt in wine. Chlorides can affect the wine's taste and also indicate the quality of the grapes used in making the wine.

6-Free sulfur dioxide: It is a measure of the amount of sulfur dioxide added to wine as a preservative. It helps prevent spoilage and oxidation, and its levels can indicate the quality of the winemaking process.

7-Total sulfur dioxide: It is the sum of the free and bound sulfur dioxide in wine. It can influence the wine's taste, aroma, and overall quality.

8-Density: It is a measure of the wine's mass per unit volume. It can indicate the wine's alcohol content, sweetness, and body.

9-pH: It is a measure of the acidity of wine. It can influence the wine's taste, color, and stability.

10-Sulphates: It is a measure of the amount of sulphates in wine, which act as antioxidants and preservatives. Sulphates can affect the wine's taste and aroma.

11-Alcohol: It is a measure of the percentage of alcohol in wine. It can influence the wine's body, flavor, and overall quality.

12-Quality: It is the output variable, which is a measure of the quality of wine on a scale of 0 to 10. It is influenced by various physicochemical properties of wine, as mentioned above.

When working with datasets, it is common to encounter missing data, which can adversely affect model performance. In the wine quality dataset, missing data can occur in any of the 12 features. There are various techniques to handle missing data during the feature engineering process, including:

1-Deletion: This technique involves deleting rows or columns that contain missing data. It is the simplest approach but can result in a loss of information and reduced sample size.

2-Mean/Median/Mode Imputation: In this technique, missing values are replaced with the mean, median or mode of the respective feature. It is a straightforward approach, but it assumes that the missing data is missing at random (MAR) and can introduce bias in the data.

3-K-Nearest Neighbor Imputation: This technique involves finding the k-nearest neighbors of the missing data point and imputing the missing value with the mean/median/mode of the k-nearest neighbors. It is a more sophisticated approach than mean/median/mode imputation but can be computationally intensive.

4-Regression Imputation: This technique involves using a regression model to predict the missing values based on the values of the other features. It can be a powerful approach but assumes that there is a linear relationship between the features and can be sensitive to outliers.

```python
# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
# features. What is the minimum number of principal components required to explain 90% of the variance in
# the data?

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, OneHotEncoder

df = pd.read_csv('winequalityN.csv')
onehot = OneHotEncoder()
type_onehot = onehot.fit_transform(df[["type"]]).toarray()


scaler = StandardScaler()
num_features = scaler.fit_transform(df.drop(["type", "quality"], axis=1))

X = pd.concat([pd.DataFrame(type_onehot, columns=["red", "white"]), pd.DataFrame(num_features, columns=df.columns[1:-1])], axis=1)
X.isna().sum()
X.dropna(inplace=True)
pca = PCA()
pca.fit(X)

import matplotlib.pyplot as plt
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()

cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
n_components = np.argmax(cumulative_variance_ratio >= 0.9) + 1
print("Minimum number of principal components required to explain 90% of the variance:", n_components)

```
